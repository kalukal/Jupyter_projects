```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x='Gender')
for p in ax.patches:
    ax.annotate(p.get_height(),(p.get_x()+p.get_width()/2,p.get_height()),ha='center', va='baseline',fontsize=15,fontstyle='italic')
plt.title('Gender Distribution')
```

```python
#looking at the avg score of student based on their parents education qualification
```

```python
df.head()
```

```python
df['ParentEduc'].value_counts()
```

```python
fig,ax=plt.subplots(3,1,figsize=(10,15))
sns.boxplot(data=df,x='ParentEduc',y='MathScore',ax=ax[0],showmeans=True,meanline=True,meanprops={'color':'red'})
sns.boxplot(data=df,x='ParentEduc',y='ReadingScore',ax=ax[1],showmeans=True,meanline=True,meanprops={'color':'red'})
sns.boxplot(data=df,x='ParentEduc',y='WritingScore',ax=ax[2],showmeans=True,meanline=True,meanprops={'color':'red'})
plt.suptitle('Relationship between parent education and marks',fontsize=30)
```

```python
data=df.groupby('ParentEduc')[['MathScore','ReadingScore','WritingScore']].mean().round()
data
```

```python
plt.figure(figsize=(5,6))
sns.heatmap(data,annot=True)
plt.title("Relationship between parent Education and marks(Heat Map)")
```

```python
#Distribution of ethnic GROUP
```

```python
df.head()
```

```python
new=df['EthnicGroup'].value_counts(normalize=True)*100
new=new.round(2)
new=new.to_frame()
new=new.reset_index()
new
```

```python
plt.pie(new['EthnicGroup'],labels=new['index'],autopct='%1.1f%%',startangle=140)
plt.title('Ethnic Group Distribution')
sns.set(style="whitegrid")
```

```python
ax=sns.barplot(data=new,x='index',y='EthnicGroup')
ax.bar_label(ax.containers[0],fmt='%.2f%%',label_type='edge',fontsize=12,color='black')
plt.title("Ethnic group Distribution")
```
