# Online order facility vs location

```python
d0=data.groupby(['online_order','location'])['name'].count()
d1=pd.DataFrame(d0)
d1=pd.pivot_table(d1,values=None,index=['location'],columns=['online_order'],fill_value=0,aggfunc=np.sum)
```

```python
pivot_melted = d1.reset_index()

# Melt the DataFrame to have 'No' and 'Yes' in a single column
melted_df = pd.melt(pivot_melted, id_vars=['location'], var_name=['order_type', 'online_order'], value_name='count')

# Create the bar plot using Seaborn
plt.figure(figsize=(14, 8))
sns.barplot(x='location', y='count', hue='online_order', data=melted_df, palette='Set2')
plt.title('Number of Restaurants Providing Online Order in Different Locations (Pivot Table)')
plt.xticks(rotation=90)
plt.show()
```
