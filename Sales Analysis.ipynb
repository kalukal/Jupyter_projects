```python
data1=data.loc[data['YEAR_ID']==2003,['COUNTRY','SALES']]
data2=data.loc[data['YEAR_ID']==2004,['COUNTRY','SALES']]
data3=data.loc[data['YEAR_ID']==2005,['COUNTRY','SALES']]
```

```python
new_data1=data1.groupby('COUNTRY').sum('SALES').sort_values('SALES',ascending=False).head(5)
new_data2=data2.groupby('COUNTRY').sum('SALES').sort_values('SALES',ascending=False).head(5)
new_data3=data3.groupby('COUNTRY').sum('SALES').sort_values('SALES',ascending=False).head(5)
```

```python
#Creating the outline structure using subplot()
f,(sub1,sub2,sub3)=plt.subplots(nrows=1,ncols=3,figsize=(20,10))
#suptitle() function to set the title 
f.suptitle('Country based SALES(TOP-5) chart',font='bold',size=20)

## first bar chart
sub1.bar(new_data1.index,new_data1['SALES'])
sub1.set_title('2003',font='bold',color='black',size=15)
# for loop to put datalabel on first bar chart by using text() function
for i, v in enumerate(new_data1['SALES']):
    sub1.text(i, v+1, v, ha='center',va='bottom',fontsize=11,color='black')
    
## second bar chart
sub2.bar(new_data2.index,new_data2['SALES'],color='red')
sub2.set_title('2004',font='bold',color='black',size=15)
# for loop to put datalabel on first bar chart by using text() function
for i, v in enumerate(new_data2['SALES']):
    sub2.text(i, v+1, v, ha='center',va='bottom',fontsize=11,color='black')
    
## Third bar chart
sub3.bar(new_data3.index,new_data3['SALES'],color='green')
sub3.set_title('2005',font='bold',color='black',size=15)
# for loop to put datalabel on first bar chart by using text() function
for i, v in enumerate(new_data3['SALES']):
    sub3.text(i, v+1, v, ha='center',va='bottom',fontsize=11,color='black')
```

```python
gdata=data[['YEAR_ID','QTR_ID','SALES']]
```

```python
new_gdata=gdata.groupby(['YEAR_ID','QTR_ID']).sum('SALES').reset_index()
pivot_gdata=new_gdata.pivot(index='QTR_ID',columns='YEAR_ID',values='SALES').fillna(0).astype('int')
print(pivot_gdata)
fig, ax = plt.subplots(figsize=(8, 6))
fig.suptitle('SALES PER QUATERS')
pivot_gdata.plot(kind='bar',ax=ax)
ax.set_ylim(0,3000000)
ax.set_ylabel('Value(Lakhs)')
ax.set_xlabel('Quaters')
```

```python
data2=data[['YEAR_ID','PRODUCTCODE']]
data3=data2.groupby(by=['YEAR_ID','PRODUCTCODE']).size().reset_index(name='Frequency')
A=data3.loc[data3['YEAR_ID']==2003,['PRODUCTCODE','Frequency']].sort_values('Frequency',ascending=False).head(5).set_index('PRODUCTCODE')
B=data3.loc[data3['YEAR_ID']==2004,['PRODUCTCODE','Frequency']].sort_values('Frequency',ascending=False).head(5).set_index('PRODUCTCODE')
C=data3.loc[data3['YEAR_ID']==2005,['PRODUCTCODE','Frequency']].sort_values('Frequency',ascending=False).head(5).set_index('PRODUCTCODE')
```

```python
f,ax=plt.subplots(1,3,figsize=(15,5))
f.suptitle('[TOP-5]MOST SOLD PRODUCT')
A.plot(kind='bar',ax=ax[0],title='2003',legend=False,color='blue')
B.plot(kind='bar',ax=ax[1],title='2004',legend=False,color='red')
C.plot(kind='bar',ax=ax[2],title='2005',legend=False,color='green')
```
