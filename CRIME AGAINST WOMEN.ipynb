```python
data1=data.loc[:,['STATE/UT','Year','Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']]
data1
```

```python
list=['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']
new_data1=data1.groupby(by=['STATE/UT','Year'],axis=0)[list].sum()
```

```python
new_data2=DataFrame(new_data1)
new_data2
data3=new_data2.reset_index()
data3
```

```python
Year_widget = widgets.Dropdown( # Creating Dropdown widget with Year as option.
    options=data3['Year'].unique(),
    value=data3['Year'].iloc[0],
    description='Select Year:',
    disabled=False,
)
```

```python
def update_data(Year):# function to give the given data
    DATA=data3[data3['Year']==Year].head(5).sort_values(by='Rape',ascending=False)# Getting top 5 values of desired data
    fig,ax=plt.subplots(nrows=3,ncols=2,figsize=(10,10))# Creating Figure and Subplot
    sns.barplot(data=DATA,x='Rape',y='STATE/UT',orient='h',ax=ax[0,0]) # plot all the Graph in their respective axes.
    sns.barplot(data=DATA,x='Kidnapping and Abduction',y='STATE/UT',orient='h',ax=ax[0,1])
    sns.barplot(data=DATA,x='Dowry Deaths',y='STATE/UT',orient='h',ax=ax[1,0])
    sns.barplot(data=DATA,x='Assault on women with intent to outrage her modesty',y='STATE/UT',orient='h',ax=ax[1,1])
    sns.barplot(data=DATA,x='Importation of Girls',y='STATE/UT',orient='h',ax=ax[2,0])
    sns.barplot(data=DATA,x='Cruelty by Husband or his Relatives',y='STATE/UT',orient='h',ax=ax[2,1])
    plt.subplots_adjust(wspace=.6,hspace=0.3) #Adjust the horizontal and vertical space.
    for x in range(0,3): # fucntion to sex all label to ''.
        for y in range(0,2):
            ax[x,y].set_xlabel('')
            ax[x,y].set_ylabel('')
    ax[0,0].set_title('Rape')# Setting titles for each plots.
    ax[0,1].set_title('Kidnapping and Abduction')
    ax[1,0].set_title('Dowry Deaths')
    ax[1,1].set_title('Assault on Women')
    ax[2,0].set_title('Importation of Girls')
    ax[2,1].set_title('Cruelty by husband or his Relatives')
    '''
       for m in range(0,3):
        for n in range(0,2):   # This is a func which can be used to plot labels but it was becoming messy so I choose not to use it.
            for x in ax[m, n].patches:
                ax[m, n].text(x.get_width(), x.get_y() + x.get_height() / 2, int(x.get_width()), fontsize=10, color='black', ha='right')
    '''
```

```python
widgets.interact(update_data, Year=Year_widget); #interact func is used to map the widget data to the dataframe.
```
