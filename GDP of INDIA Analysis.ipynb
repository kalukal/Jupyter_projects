```python
def percentage(data):
    growth=[]
    for x in range(0,len(data)-1):
        growth.append(((data[x+1]-data[x])/data[x])*100)
    grow=[math.ceil(x) for x in growth]
    grow.insert(0,0)
    return grow
        
        
```

```python
gdp_per=percentage(Data3['GDP'])
Data3['per_gdp']=gdp_per
Data3
```

```python
capita_per=percentage(Data3['GDP_per_capita'])
Data3['per_growth']=capita_per
Data3.drop('groth_per',axis=1)
```

```python
ax=plt.figure(figsize=(10,5))
plt.plot(Data3['Year'],Data3['GDP_per_capita'],label='GDP growth percent')
plt.plot(Data3['Year'],Data3['groth_per'],label='PER CAPITA growth percent')
plt.legend(loc='upper left')
plt.title('DIFF BETWEEN GDP AND PER CAPITA GROWTH')
```
