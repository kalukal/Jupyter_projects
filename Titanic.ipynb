```python
sur_data=mydata[mydata['Survived']==1]
type(sur_data)
```

```python
ax=sns.countplot(data=mydata,x='Pclass')
for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2,p.get_height(),(p.get_height()),ha='center',va='bottom')
plt.title('No.of people in different classes')
plt.savefig("C:\\Users\\rawat\\Desktop\\picture\\Total no of people in different class")
```

```python
#Analysis of Avg age of people on board based on sex.
sns.kdeplot(data=sur_data,x='Age',hue='Sex',fill=True)
plt.xlim(0)
plt.title('Graph to show the denstiy of people on board by sex')
plt.savefig("C:\\Users\\rawat\\Desktop\\picture\\Desity of people divided by sex")
```

```python
sns.displot(data=mydata,x='Age',hue='Pclass',kind="kde",fill=True)
plt.title('Graph to show different Age group of people in different classes')
plt.savefig("C:\\Users\\rawat\\Desktop\\picture\\Desity of people divided by Class")
```

```python
#percentage of people on board per sex.
sex_data=[mydata[mydata['Sex']=='male']['PassengerId'].count(),mydata[mydata['Sex']=='female']['PassengerId'].count()]
plt.pie(sex_data,labels=['male','female'],autopct='%1i%%',textprops = {'color': 'w'})
plt.legend(['male','female'])
plt.title('People divided by sex')
plt.savefig("C:\\Users\\rawat\\Desktop\\picture\\Percentage of people on board")
```

```python
#No. of people on board in different Pclass of different sex.
sns.set_style('white')
ax=sns.countplot(x='Pclass',data=mydata,hue='Sex')
plt.title('No of Men vs Women per Class')
for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2,p.get_height(),p.get_height(),ha='center',va='bottom')
    plt.savefig("C:\\Users\\rawat\\Desktop\\picture\\No of people on board per class")
```

```python
#Creating the Child data to see how many no of people below 16 were on board
def gender(data):
    age,sex=data
    if age<16:
        return 'Child'
    else:
        return sex
```

```python
mydata['gender']=mydata[['Age','Sex']].apply(gender,axis=1)
```

```python
# Casuality Ratio between MALE, FELMALE, CHILD
ax=sns.countplot(x=mydata[mydata['Survived']==0]['gender'])
plt.title('Casuality Ratio between Male,Female,Child')
for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2,p.get_height(),p.get_height(),ha='center',va='bottom')
plt.savefig("C:\\Users\\rawat\\Desktop\\picture\\Casuality ratio between Male,female,child")
```

```python
# Survival ratio between male , female, child
ax=sns.countplot(x=mydata[mydata['Survived']==1]['gender'])
plt.title('Survived Ratio between Male,Female,Child')
for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2,p.get_height(),p.get_height(),ha='center',va='bottom')
plt.savefig("C:\\Users\\rawat\\Desktop\\picture\\Survival ratio between male female and child")
```
